```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('bmh')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data (downloaded from Kaggle - https://www.kaggle.com/datasets/danielgrijalvas/movies?resource=download)

df = pd.read_csv('/Users/chenhaoxin/Documents/DA Projects/Python - Correlation Movie Dataset/movies.csv.zip')
```

```python
# Check data

df.head()
```

```python
# Remove missing values

df = df.dropna()

# Check for percentage of missing data

for col in df.columns:
    perc_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, perc_missing))
    
```

```python
# Data types for columns

df.dtypes
```

```python
# Change the data type of particular columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct year column

NewDataFrame = df["released"].str.split(" ",n = 3,expand = True) 
NewDataFrame.rename(columns = {0: "Month", 1: "Day",2:"Year",3:"Country"}, inplace = True)
df['year_correct'] = NewDataFrame['Year']

""" Alternative way
df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(float)
df['year_correct'].fillna(0, inplace = True)
df['year_correct'] = df['year_correct'].astype(int)
"""

df.head()
```

```python
# Order data by Gross (descending) to have a better insight

df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Display every rows

pd.set_option('display.max_rows', None)

df.head()
```

```python
# Drop duplicates

df.drop_duplicates()

df.head()
```

```python
# Explore correlations
```

```python
# Scatter plot w/ budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings ($)')
plt.ylabel('Budget for Film ($)')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn for visualisation

sns.regplot(x = 'budget', y = 'gross', data = df, line_kws = {"color":'orange'})
```

```python
# Check correlation

df.corr(method = 'pearson')

# budget & gross have highest corr: 0.74
# gross & votes second highest: 0.61
```

```python
corr_matrix = df.corr(method = 'pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Create numeric representation of company so it can be included in the Correlation Matrix

df_numerized = df

for name in df_numerized.columns:
    if(df_numerized[name].dtype == 'object'):
        df_numerized[name] = df_numerized[name].astype('category') # change column type
        df_numerized[name] = df_numerized[name].cat.codes
        
df_numerized

df.head()
```

```python
corr_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(corr_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
corr_mat = df_numerized.corr()

corr_pairs = corr_mat.unstack()

corr_pairs.head(15)
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head(15)
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
