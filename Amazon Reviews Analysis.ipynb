## Task 3 - Perform Data Cleaning On Text

```python
# Define a function to clean up all messages - remove punctuation & stopwords

import string
string.punctuation

# View the punctuations
print(string.punctuation)

import nltk # Natural Language tool kit 
nltk.download('stopwords')

# Download stopwords Package
from nltk.corpus import stopwords
stopwords.words('english')

# View the stopwords
print(stopwords.words('english'))

# Function to perform data cleaning
def message_cleaning(message):
    punc_removed = [c for c in message if c not in string.punctuation]
    punc_removed_join = ''.join(punc_removed)
    punc_removed_join_clean = [word for word in punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return punc_removed_join_clean
```

```python
reviews_df_clean = reviews_df['verified_reviews'].apply(message_cleaning)
```

```python
# View the first five Amazon reviews
print(reviews_df['verified_reviews'][0]) 
print(reviews_df['verified_reviews'][1]) 
print(reviews_df['verified_reviews'][2]) 
print(reviews_df['verified_reviews'][3]) 
print(reviews_df['verified_reviews'][4]) 
```

```python
# View the cleaned version of the five reviews

print(reviews_df_clean[0])
print(reviews_df_clean[1])
print(reviews_df_clean[2])
print(reviews_df_clean[3])
print(reviews_df_clean[4])
```
